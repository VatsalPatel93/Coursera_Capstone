Let's define a function to create a function to get nearby venues in each postal codes.

```python
def getNearbyVenues(postal_codes, names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for postal_code, name, lat, lng in zip(postal_codes, names, latitudes, longitudes):
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            postal_code,
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['PostalCode',
                  'Borough', 
                  'PC Latitude', 
                  'PC Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

We can now call the function above to get details of venues within the postal codes.

```python
borough_venues = getNearbyVenues(postal_codes=borough_data['PostalCode'],
                                 names=borough_data['Borough'],
                                 latitudes=borough_data['Latitude'],
                                 longitudes=borough_data['Longitude'])
```

We can now use one hot encoding to convert categorical attribute categories into a multiple numerical attrubutes

```python
# one hot encoding
borough_onehot = pd.get_dummies(borough_venues[['Venue Category']], prefix="", prefix_sep="")

# add Postal Code and Borough column back to dataframe
borough_onehot['PostalCode'] = borough_venues['PostalCode'] 
borough_onehot['Borough'] = borough_venues['Borough'] 

# move PostalCode and Borough columns to the first and second column respectively
fixed_columns = list(borough_onehot.columns[-2:]) + list(borough_onehot.columns[:-2])
borough_onehot = borough_onehot[fixed_columns]

borough_onehot.head()
```

Let's group the data by Postal Code and Borough ans take a mean of frequencies for each category of venue

```python
borough_grouped = borough_onehot.groupby(['PostalCode','Borough']).mean().reset_index()
borough_grouped.shape
```

Let's see what top 5 venue categories are in each (Postal Code, Borough)

```python
num_top_venues = 5

for pc, borough in zip(borough_grouped['PostalCode'],borough_grouped['Borough']):
    print("----"+pc+ " : "+borough+"----")
    temp = borough_grouped[(borough_grouped['PostalCode'] == pc) & (borough_grouped['Borough'] == borough)].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[2:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

Let's convert this into a dataframe we can use for analysis. We will first write a function that returns top venue categories and then get top 10 venues for each Postal Code, Borough

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[2:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

Now let's create the new dataframe and display the top 10 venues for each neighborhood.

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['PostalCode','Borough']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
borough_venues_sorted = pd.DataFrame(columns=columns)
borough_venues_sorted['PostalCode'] = borough_grouped['PostalCode']
borough_venues_sorted['Borough'] = borough_grouped['Borough']

borough_grouped.iloc[0,:].iloc[2:].sort_values(ascending=False).index.values[0:10]
for ind in np.arange(borough_grouped.shape[0]):
    borough_venues_sorted.iloc[ind, 2:] = return_most_common_venues(borough_grouped.iloc[ind, :], num_top_venues)

borough_venues_sorted.head()
```

## Clustering
Let's Cluster PostalCodes and boroughs into  clusters.

```python
# set number of clusters
kclusters = 5

borough_grouped_clustering = borough_grouped.drop(['PostalCode','Borough'], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, algorithm='full',random_state=0).fit(borough_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels
borough_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

borough_merged = borough_data

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
borough_merged = borough_merged.join(borough_venues_sorted.set_index(['PostalCode','Borough']), on=['PostalCode', 'Borough'])

borough_merged.head() # check the last columns!
```

Finally, let's visualize the resulting clusters

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, postal_code, borough, cluster in zip(borough_merged['Latitude'], 
                                                   borough_merged['Longitude'], 
                                                   borough_merged['PostalCode'], 
                                                   borough_merged['Borough'], 
                                                   borough_merged['Cluster Labels']):
    label = folium.Popup(postal_code + ',' + borough + ': Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
